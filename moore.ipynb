```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
with open('datasets\moore.csv.txt') as f:
    for line in f:
      Y.append(line.split("\t")[1])
      X.append(line.split("\t")[2])  
```

```python
Y[-1] = '7,200,000,000'
Y[-3] = '1,750,000,000'
for index in range(len(Y)):
    if "[" in Y[index]:
        Y[index] = Y[index][:-4]
    Y[index] = int(Y[index].replace(',', ''))
    
Y = np.array(Y)
```

```python
X[5] = '1975'
for index in range(len(X)):
    if "[" in X[index]:
        X[index] = X[index][:-4]
    X[index] = int(X[index])

X = np.array(X)
```

```python
Y = np.log(Y)
```

```python
plt.scatter(X,Y);
```

```python
denominator = np.mean(X**2)-(np.mean(X)**2)
weight_0 = (np.mean(X*Y) - np.mean(X)*np.mean(Y))/denominator
weight_1 = (np.mean(X**2)*np.mean(Y) - np.mean(X)*np.mean(X*Y))/denominator
r_squared = 1 - (np.sum((Y-(weight_0*X+weight_1))**2)/np.sum((Y-np.mean(Y))**2))
print(r_squared)
plt.scatter(X, Y)
plt.plot(X, (weight_0*X+weight_1), c='red')
plt.show()
```
