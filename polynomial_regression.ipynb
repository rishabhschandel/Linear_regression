```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
with open('datasets/data_poly.csv') as f:
    for line in f:
        x = float(line.split(',')[0])
        y = float(line.split(',')[1])
        X.append([1, x, x*x])
        Y.append(y)
X=np.array(X)
Y=np.array(Y)

```

```python
plt.figure()
plt.scatter(X[:,1], Y);
```

```python
weights = np.linalg.solve(np.dot(X.T,X), np.dot(X.T,Y))
```

```python
weights
```

```python
Yhat = np.dot(X,weights)
```

```python
d1 = Yhat-Y
d2 = Yhat - Y.mean()
r_squared = 1 - d1.dot(d1)/d2.dot(d2)
```

```python
r_squared
```
