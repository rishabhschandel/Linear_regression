```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
df = pd.read_csv('datasets/data_2d.csv', header=None)
df.head()
```

```python
X = df.iloc[:,:2].values
X = np.concatenate((np.ones((100,1)),np.random.randn(100,1), X), axis = -1)
X
```

```python
Y = df.iloc[:,2].values
Y.shape
```

```python
weights = np.linalg.solve(np.dot(X.T,X),np.dot(X.T,Y))
weights.shape
```

```python
Yhat = np.dot(X,weights) 
```

```python
d1 = Yhat - Y
d2 = Yhat - Y.mean()
```

```python
r_squared = 1 - d1.dot(d1)/d2.dot(d2)
r_squared
```
